```python
def analyze_student_performance(student_data):
    return f"""
Student Performance Report

Name: {student_data['name']}

Subjects: {student_data['subjects']}
Attendance: {student_data['attendance']}
Trend: {student_data['trend']}

Summary:
- Student shows mixed performance across subjects.
- Attendance is moderate and may be affecting consistency.
- Improvement trend is visible but needs strengthening.

Recommendations:
- Focus more on weaker subjects daily
- Improve revision schedule
- Maintain consistent attendance
- Practice past papers regularly
"""
```

```python
student_data = {
    "name": "Ali",
    "subjects": {"Math": 75, "Science": 82, "English": 68},
    "attendance": "85%",
    "trend": "slightly improving"
}

print(analyze_student_performance(student_data))
```

```python
import streamlit as st

st.title("🎓 Student Performance Analyzer")

# Input fields
name = st.text_input("Student Name")

math = st.number_input("Math Score", 0, 100)
science = st.number_input("Science Score", 0, 100)
english = st.number_input("English Score", 0, 100)

attendance = st.slider("Attendance (%)", 0, 100, 80)

trend = st.selectbox("Performance Trend", [
    "Improving",
    "Declining",
    "Stable",
    "Inconsistent"
])

# Analysis function (offline version)
def analyze(name, math, science, english, attendance, trend):
    avg = (math + science + english) / 3

    result = f"""
## 📊 Report for {name}

### 🎯 Scores
- Math: {math}
- Science: {science}
- English: {english}
- Average: {avg:.2f}

### 📅 Attendance
- {attendance}%

### 📈 Trend
- {trend}

---

## 🧠 Analysis

"""

    if avg >= 80:
        result += "- Excellent performance overall.\n"
    elif avg >= 60:
        result += "- Average performance, room for improvement.\n"
    else:
        result += "- Weak performance, needs strong improvement.\n"

    if attendance < 75:
        result += "- Low attendance may be affecting performance.\n"

    if trend == "Declining":
        result += "- Performance is dropping; immediate action needed.\n"

    if math < 60:
        result += "- Focus needed in Math.\n"

    if science < 60:
        result += "- Focus needed in Science.\n"

    if english < 60:
        result += "- Focus needed in English.\n"

    result += """
---

## ✅ Recommendations
- Study daily for 2–3 hours
- Revise weak subjects
- Improve attendance
- Practice past papers
"""

    return result


# Button
if st.button("Generate Report"):
    report = analyze(name, math, science, english, attendance, trend)
    st.markdown(report)
```
